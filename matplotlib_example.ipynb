```python
import matplotlib.pyplot as plt #para o Python importar da biblioteca matplotlib a parte de criar/plotar gráficos com o Python (pyplot) e o "plt" é a abreviação.
```

Criando um gráfico com uma planilha de vendas

---

```python
meses = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez']
valores = [497, 5369, 2958, 1187, 3864, 850, 9584, 5260, 2789, 1100, 7853, 15879]

plt.plot(meses, valores, color='green') #criando/plotando o gráfico e guardando na memória

plt.ylim(0, 20000) #para alterar o limite dos valores minimo e máximo

plt.title("Faturamento no ano de 2000", color='darkgreen') #adicionando um título ao gráfico

plt.xlabel("Meses (eixo x)", color='green') #adicionando significados nos eixos 'x' e 'y'
plt.ylabel("Faturamento (eixo y)", color='green')

plt.xticks(rotation=45) #para rotação das palavras do eixo 'x'

plt.show() #mostrando o gráfico
```
